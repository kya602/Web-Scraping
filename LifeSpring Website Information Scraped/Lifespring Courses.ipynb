```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"}
```

```python
#Global List
courseList = []
```

```python
def getEntries(soup):
    
    for courses in soup.find_all('div', class_='course-loop-info'):
        
        lesson = courses.find('div', class_='course-loop-meta-item course-loop-meta-lesson')
        duration = courses.find('div', class_='course-loop-meta-item course-loop-meta-duration')
        
        if lesson is not None:
            lesson = lesson.find('div', class_='meta-value').text.strip()
        else:
            lesson = None
            
        if duration is not None:
            duration = duration.find('div', class_='meta-value').text.strip()
        else:
            duration = None
        
        entry = {
        'Title' : courses.find('h2', class_ = 'course-loop-title course-loop-title-collapse-2-rows').a.text,

        'Description' : str(courses.find('div', class_='course-loop-excerpt course-loop-excerpt-collapse-2-rows').p).strip('</p>'),

        'Price(৳)' : courses.find('span', class_='woocommerce-Price-amount amount').bdi.text.strip('৳\xa0.'),
            
        'ClassNo' : lesson,
            
        'Total Duration' : duration
        
        }
        courseList.append(entry)
        
    return
    
```

```python
base_url = 'https://www.lifespringint.com/courses/'

html = requests.get(base_url)
soupB = BeautifulSoup(html.content,'html.parser')
page_no = len(soupB.find_all('a',class_='page-numbers'))

for i in range(1, page_no+1):
    url = base_url + f'page/{i}'
    page = requests.get(url).content
    soup = BeautifulSoup(page,'html.parser')
    getEntries(soup)
```

```python
df = pd.DataFrame(courseList)
df
```
