```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Global List
blogList = []
```

```python
def getEntries(soup):
    
    for blogs in soup.find_all('div', class_='post-caption'):
        
        entry = {
            
        'Title' : blogs.find('h3', class_='post-title').text.strip(),

        'Description' : str(blogs.find('div', class_='post-excerpt').p).strip('</p>'),

        'Date Published' : blogs.find('div', class_='post-date').find('span', class_='meta-value').text.strip(),
            
        'Views' : blogs.find('div', class_='post-view').find('span', class_='meta-value').text.strip(),
            
        'URL (link)' : blogs.find('h3', class_='post-title').a['href']
        
        }
        blogList.append(entry)
        
    return
    
```

```python
base_url = 'https://www.lifespringint.com/blogs/'

html = requests.get(base_url)
soupB = BeautifulSoup(html.content,'html.parser')
page_no = len(soupB.find_all('a',class_='page-numbers'))

for i in range(1, page_no+1):
    url = base_url + f'page/{i}'
    page = requests.get(url).content
    soup = BeautifulSoup(page,'html.parser')
    getEntries(soup)
```
